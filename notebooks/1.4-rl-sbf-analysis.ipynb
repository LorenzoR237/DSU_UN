# 1.4 — Small Business Focus Certified Business Analysis

## Purpose

Profile the 446 Delaware businesses that have completed Small Business Focus (SBF) and related supplier-diversity certifications. This is not a hypothesis test — it's an exploratory analysis that surfaces the actual population of DE SMEs positioned for anchor procurement.

## Why this matters

H3 established that ~7% of Delaware's 46,216 active business licenses are in sectors aligned with anchor procurement (Manufacturing, Wholesale, Professional Services) — roughly 3,200 eligible firms. This notebook asks: **of those eligible, how many have actually completed the certification legwork required to bid on most state, federal, and corporate procurement opportunities?**

The gap between "operating in an eligible sector" and "certified to bid" is a specific, addressable bottleneck the case's solution can target directly.

## Questions this notebook answers

1. **Who is certified?** Industry mix, certification-agency distribution (DelDOT, NMSDC, WBENC, SBAVet, etc.), overlap between certifications.
2. **Where are they?** County and ZIP distribution. Is the certified population itself urban-biased?
3. **How do they compare to the overall license base?** Cross-reference by NAICS or Business Activity to test whether certified businesses skew toward certain sectors more than the eligible pool does.
4. **What's the certification rate?** Ratio of certified firms to procurement-eligible firms, as the sharpest actionable finding.

## Not a hypothesis test — but still pre-committed

This notebook doesn't test a falsifiable claim. Instead, it commits to reporting:

- Total count and share of certified businesses
- Certification-agency breakdown
- Industry distribution vs. the eligible base
- Geographic distribution by county
- Any striking pattern in the certification-overlap matrix

If the analysis surfaces something that *would* have been a hypothesis, it gets added to the surprise log and potentially spun into a follow-up test.

## Known limitations

1. **Self-reported data.** Businesses appear in the SBF directory because they voluntarily applied. Non-appearance doesn't prove a business is unqualified — it may mean they haven't applied, don't know the program exists, or have applied to a different certification (federal SBA 8(a), for instance) not reflected here.
2. **Certification ≠ active bidding.** A certified firm may not actually be pursuing procurement opportunities. This analysis measures *readiness*, not *engagement*.
3. **Directory may lag reality.** The file is dated 2025–2026 and may include businesses that have since ceased operations.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from src import config
import pandas as pd

# Load the SBF directory
sbf = pd.read_csv(config.SBF_RAW, low_memory=False)

print(f"Rows: {len(sbf)}")
print(f"Columns: {len(sbf.columns)}")
print()
print("Key fields preview:")
print(sbf[["Name", "City", "State", "Zip Code"]].head(5).to_string(index=False))
```

```python
# --- Geographic distribution of SBF-certified businesses -----------------
print("Certified businesses by state (headquartered):")
print(sbf["State"].value_counts().head(10))
print()

# How many are headquartered in DE?
de_hq = sbf[sbf["State"] == "Delaware"]
print(f"Headquartered in Delaware: {len(de_hq):,}  ({len(de_hq)/len(sbf):.1%})")
print(f"Headquartered out of state: {len(sbf) - len(de_hq):,}  ({1 - len(de_hq)/len(sbf):.1%})")
print()

# Of the DE-HQ ones, which cities?
print("Top 15 cities for DE-headquartered certified businesses:")
print(de_hq["City"].str.strip().str.upper().value_counts().head(15))
print()

# ZIP distribution for DE-HQ firms — map to our county-FIPS convention
# Normalize ZIP (same approach as licenses notebook: strip non-digits, first 5)
de_hq = de_hq.copy()
de_hq["zip5"] = (
    de_hq["Zip Code"].astype(str)
    .str.strip()
    .str.replace(r"\D", "", regex=True)
    .str[:5]
)
print(f"Distinct DE ZIPs among certified businesses: {de_hq['zip5'].nunique()}")
print()
print("Top 15 ZIPs:")
print(de_hq["zip5"].value_counts().head(15))
```

```python
# --- Certification profile (DE-headquartered firms only) -----------------
de_hq = sbf[sbf["State"] == "Delaware"].copy()
print(f"Analyzing {len(de_hq)} DE-headquartered SBF-certified businesses")
print()

# Certification columns — boolean-style flags
cert_columns = ["DelDOT", "MDDOT", "NMSDC", "Others", "PAUCP",
                "SBAVet", "WBENC", "City of Wilmington"]

# What does a "certified" value actually look like in these columns?
print("Sample values in certification columns (first 3 rows, all cert columns):")
print(de_hq[cert_columns].head(3))
print()

# Count how each column is populated
print("Non-null / non-zero counts per certification column:")
for col in cert_columns:
    # Values might be 0/1, Yes/No, NaN/something, etc. Check what's there.
    val_counts = de_hq[col].value_counts(dropna=False)
    print(f"\n{col}:")
    print(val_counts.head())
```

```python
# --- Stacked-certification overlap analysis ------------------------------
cert_columns = ["DelDOT", "MDDOT", "NMSDC", "PAUCP", "SBAVet",
                "WBENC", "City of Wilmington"]
# Exclude "Others" — all nulls, no signal

# Convert YES/NaN to 1/0 for arithmetic
for col in cert_columns:
    de_hq[col] = (de_hq[col] == "YES").astype(int)

# Count stacked certs per firm
de_hq["n_stacked_certs"] = de_hq[cert_columns].sum(axis=1)

print(f"Total DE-HQ SBF-certified firms: {len(de_hq)}")
print()
print("Distribution of stacked certifications per firm:")
print(de_hq["n_stacked_certs"].value_counts().sort_index())
print()

# Firms with at least one stacked cert
with_stacked = de_hq[de_hq["n_stacked_certs"] > 0]
print(f"Firms with ≥1 stacked certification: {len(with_stacked)} ({len(with_stacked)/len(de_hq):.1%})")
print(f"Firms with SBF only (no stacks):     {len(de_hq) - len(with_stacked)} ({1 - len(with_stacked)/len(de_hq):.1%})")

print()
print("Firms with 2+ stacked certifications:")
multi = de_hq[de_hq["n_stacked_certs"] >= 2]
if len(multi) > 0:
    print(multi[["Name", "City"] + cert_columns + ["n_stacked_certs"]].to_string(index=False))
else:
    print("  None — no firm holds 2+ stacked certifications.")
```

```python
# --- Industry profile of DE-HQ certified firms ---------------------------
industry_columns = ["Architecture and Engineering", "Construction",
                    "Manufacturing", "Retail", "Service", "Wholesale"]

# First, check the format of these columns
print("Sample values in industry columns (first 3 rows):")
print(de_hq[industry_columns].head(3))
print()

# Count non-null per column
print("Non-null / non-zero counts per industry column:")
for col in industry_columns:
    val_counts = de_hq[col].value_counts(dropna=False)
    print(f"\n{col}:")
    print(val_counts.head())
```

```python
# --- Certification rate: certified firms vs. H3-eligible pool ------------
# From H3, DE has ~46,216 active business licenses. Of those, roughly 
# 7% are in "value-chain-aligned" sectors (Manufacturing + Wholesale + 
# Professional Services). That's the H3 eligible pool.

# H3 sector counts (from notebook 2.0)
H3_MANUFACTURING = 601
H3_WHOLESALE = 1932
H3_PROFESSIONAL_SVCS = 692
H3_ELIGIBLE_TOTAL = H3_MANUFACTURING + H3_WHOLESALE + H3_PROFESSIONAL_SVCS

# SBF-certified firms (DE-HQ) by flagged industry
sbf_manufacturing = (de_hq["Manufacturing"] == "YES").sum()
sbf_wholesale = (de_hq["Wholesale"] == "YES").sum()
sbf_arch_eng = (de_hq["Architecture and Engineering"] == "YES").sum()
# "Architecture and Engineering" is our rough proxy for Professional Services
# in the SBF schema — imperfect but defensible

print("=" * 70)
print("CERTIFICATION RATE: SBF-certified vs H3 value-chain-aligned pool")
print("=" * 70)
print()
print(f"{'Category':<30} {'H3 Pool':>10} {'Certified':>12} {'Rate':>8}")
print("-" * 70)
print(f"{'Manufacturing':<30} {H3_MANUFACTURING:>10,} {sbf_manufacturing:>12,} "
      f"{sbf_manufacturing/H3_MANUFACTURING:>8.1%}")
print(f"{'Wholesale & Distribution':<30} {H3_WHOLESALE:>10,} {sbf_wholesale:>12,} "
      f"{sbf_wholesale/H3_WHOLESALE:>8.1%}")
print(f"{'Professional Svcs (A&E proxy)':<30} {H3_PROFESSIONAL_SVCS:>10,} "
      f"{sbf_arch_eng:>12,} {sbf_arch_eng/H3_PROFESSIONAL_SVCS:>8.1%}")
print("-" * 70)
sbf_total_eligible = sbf_manufacturing + sbf_wholesale + sbf_arch_eng
print(f"{'Combined eligible':<30} {H3_ELIGIBLE_TOTAL:>10,} "
      f"{sbf_total_eligible:>12,} {sbf_total_eligible/H3_ELIGIBLE_TOTAL:>8.1%}")
print()

# Broader denominator: all 339 DE-certified vs all 46,216 active licenses
print(f"All DE-HQ SBF-certified firms: {len(de_hq)}")
print(f"All DE active business licenses: 46,216")
print(f"Share of all licensed DE firms that are SBF-certified: "
      f"{len(de_hq)/46216:.2%}")
print()

# The "procurement-ready elite" ratio
stacked_firms = (de_hq["n_stacked_certs"] > 0).sum()
print(f"Of the {len(de_hq)} SBF-certified DE firms, "
      f"{stacked_firms} ({stacked_firms/len(de_hq):.1%}) "
      f"hold at least one stacked specialty certification")
print(f"Out of 46,216 total DE licenses, that's {stacked_firms/46216:.3%} "
      f"of the business base that is fully procurement-ready")
```

```python
# --- Fig: SBF Certification Funnel (log scale, human-readable) -----------
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

stages = [
    ("Active business licenses", 46216),
    ("Operating in value-chain-aligned sectors", 3225),
    ("SBF-registered", 339),
    ("SBF-registered AND in eligible sector", 37),
    ("Hold ≥1 stacked specialty certification", 15),
    ("Hold ≥2 stacked certifications", 5),
]

labels = [s[0] for s in stages]
values = [s[1] for s in stages]

fig, ax = plt.subplots(figsize=(12, 6.5))

# Sequential palette — darkest for the tightest bottleneck
palette = ["#C6DBEF", "#9ECAE1", "#6BAED6", "#4292C6", "#2171B5", "#08306B"]

y_positions = list(range(len(stages)))
bars = ax.barh(y_positions, values, color=palette,
               edgecolor="white", height=0.68)

# Bold value + percentage labels to the right of each bar
for bar, val in zip(bars, values):
    share = val / values[0] * 100
    ax.text(bar.get_width() * 1.08,
            bar.get_y() + bar.get_height()/2,
            f"{val:,}   ({share:.2f}%)" if val < 1000 else f"{val:,}   ({share:.1f}%)",
            va="center", fontsize=11, color="#333333", fontweight="bold")

ax.set_yticks(y_positions)
ax.set_yticklabels(labels, fontsize=10.5)
ax.invert_yaxis()

# --- Log scale, but render tick labels as plain numbers ------------------
ax.set_xscale("log")
ax.set_xlim(3, 80000)

# Format major ticks as "10", "100", "1,000" (no scientific notation)
def plain_number(x, pos):
    if x >= 1:
        return f"{int(x):,}"
    return str(x)

ax.xaxis.set_major_formatter(FuncFormatter(plain_number))

# Major gridlines (at 10, 100, 1000, 10000)
ax.grid(axis="x", which="major", linestyle="-", alpha=0.25, color="#999999")
# Minor gridlines (at 20, 30, ..., 90, 200, ..., 900, etc.) — fainter
ax.grid(axis="x", which="minor", linestyle=":", alpha=0.15, color="#999999")

ax.set_axisbelow(True)

# Title only (no subtitle)
fig.suptitle(
    "Delaware's procurement-ready SME pipeline narrows to 15 firms",
    fontsize=14, fontweight="bold", x=0.02, y=0.98, ha="left", color="#051C2C")

ax.set_xlabel("Number of firms (log scale)", fontsize=10, color="#555555")

# Clean spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(axis="both", labelsize=9.5, colors="#555555")

plt.tight_layout(rect=[0, 0, 1, 0.94])

output_path = config.FIGURES / "fig00_sbf_funnel.png"
plt.savefig(output_path, dpi=200, bbox_inches="tight")
print(f"Saved: {output_path}")
plt.show()
```

## SBF Analysis Summary — Flagship Finding

### Headline

**Of Delaware's 46,216 active business licenses, 15 hold the specialty certifications needed for anchor procurement. That's 1 firm in every 3,080.**

### The funnel

| Stage | Firms | Share of base |
|---|---|---|
| Active DE business licenses | 46,216 | 100% |
| Operating in value-chain-aligned sectors (H3) | 3,225 | 7.0% |
| SBF-registered (baseline DE certification) | 339 | 0.73% |
| SBF-registered AND in value-chain-aligned sector | 37 | 0.08% |
| Hold ≥1 stacked specialty certification | 15 | 0.03% |
| Hold ≥2 stacked certifications | 5 | 0.01% |

### Where the bottleneck actually lives

This analysis reframes the case. The bottleneck is not:

- **Not capital availability** — $14M in SSBCI funds sat undeployed (external research)
- **Not physical lender access** — H1 showed distance tracks density, not race
- **Not state funding scale per se** — H5 showed $125/business/year, constant for a decade

**The bottleneck is certification completion.** 324 of 339 DE-registered SBF firms (95.6%) have not stacked any specialty certification on top of SBF. Specialty certifications are what actually unlock procurement channels: DelDOT for state transportation contracts, NMSDC for Fortune 500 minority-supplier programs, WBENC for women-owned business set-asides, SBAVet for federal veteran set-asides, PAUCP for regional DBE programs.

### The procurement-ready elite

Only 5 DE-HQ firms hold 2+ stacked specialty certifications. All five are construction or construction-adjacent:

1. **Frazer Solutions LLC** (Wilmington) — 4 stacks: DelDOT, MDDOT, NMSDC, SBAVet
2. **Dependable Construction of DE** (Wilmington) — 3 stacks: DelDOT, MDDOT, City of Wilmington
3. **JW WorkZone Supplies** (Dover) — 2 stacks: DelDOT, PAUCP
4. **On The Mark Locators, LLC** (Bear) — 2 stacks: DelDOT, PAUCP
5. **ADS Construction Services** (Wilmington) — 2 stacks: DelDOT, City of Wilmington

**All five hold DelDOT certification.** Regional transportation contracting is the only domain where Delaware has developed meaningful procurement-ready depth. Non-construction certification pipelines (manufacturing, professional services, women-owned services) are nearly empty.

### Certification rates by H3 eligible sector

| Eligible sector (from H3) | H3 firms | SBF-flagged firms | Cert rate |
|---|---|---|---|
| Manufacturing | 601 | 18 | 3.0% |
| Wholesale & Distribution | 1,932 | 15 | 0.8% |
| Professional Services (A&E proxy) | 692 | 4 | 0.6% |
| **Combined** | **3,225** | **37** | **1.1%** |

### Why this matters for the case

Unlike the structural findings from H3, H5, and H1, the certification bottleneck is addressable with a specific program in a concrete timeframe:

- **Dedicated certification-assistance FTEs** (2-3 staff embedded in DSB or SBDC)
- **Cohort-based certification sprints** (20-firm cohorts, 90-day programs)
- **State coverage of certification fees** (most specialty certs are $1K–$5K each)
- **Anchor-buyer match commitments** (pre-committed RFP invitations for certified firms)

A well-designed program could plausibly move Delaware from 15 procurement-ready firms to 150 in 24 months — a 10× improvement with a defined end-state.

### Methodology notes

- **Population:** 446 firms in the SBF directory; 339 DE-headquartered (76%). Out-of-state firms (24%) hold DE certifications for tri-state bidding — reported in appendix but not counted toward DE base.
- **Certification counts:** From the 7 specialty-certification columns (DelDOT, MDDOT, NMSDC, PAUCP, SBAVet, WBENC, City of Wilmington). The "Others" column is 100% null and was excluded.
- **Industry flags:** Data quality issue noted — Construction column is 100% null despite construction firms being present in the directory. Firms with DelDOT certification are by name construction contractors. The "Service" flag at 60% appears to be a catch-all. Industry flag counts reported directionally; NAICS-code refinement deferred given scope.
- **H3 cross-reference:** Manufacturing, Wholesale, and Professional Services counts from notebook 2.0. Architecture & Engineering treated as a proxy for Professional Services in the SBF schema.
- **Not a hypothesis test:** This notebook is exploratory profiling. The finding emerged from the data rather than from a pre-committed threshold.
